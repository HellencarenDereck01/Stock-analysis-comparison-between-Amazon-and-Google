## TECHNICAL ANALYSIS

Technical analysis involves evaluating securities by analyzing historical price data and market indicators. It focuses on identifying patterns, trends, and market signals based on price movement, without considering a company's fundamental financials.

In this analysis, several commonly used technical indicators were applied to both Amazon (AMZN) and Google (GOOGL) stock price data to assess trends, momentum, and volatility. These indicators are described below:

---

### 1. Simple Moving Average (SMA)

- **Definition:** The SMA calculates the average of a security’s price over a defined time period.
- **Purpose:** It smooths short-term price fluctuations and helps identify the overall direction of the market trend.
- **Application:** A 30-day SMA was used to observe medium-term trends in the stock prices.

---

### 2. Exponential Moving Average (EMA)

- **Definition:** The EMA is a type of moving average that gives more weight to recent price data.
- **Purpose:** It responds more quickly to price changes than the SMA, making it useful for identifying momentum shifts.
- **Application:** A 20-day EMA was applied to track short-term movements and compare them to longer-term trends.

---

### 3. Bollinger Bands

- **Definition:** Bollinger Bands consist of a moving average with upper and lower bands set at two standard deviations.
- **Purpose:** They measure price volatility and help identify overbought or oversold conditions.
- **Application:** The bands were used to monitor how far prices deviated from the mean and to anticipate potential reversals.

---

### 4. Relative Strength Index (RSI)

- **Definition:** RSI is a momentum oscillator that ranges from 0 to 100 and measures the speed and change of price movements.
- **Purpose:** It indicates overbought conditions when above 70 and oversold conditions when below 30.
- **Application:** RSI was used to identify potential trend reversals and assess the strength of price movements.

---

### 5. Moving Average Convergence Divergence (MACD)

- **Definition:** MACD is a momentum indicator that shows the relationship between two EMAs, typically the 12-day and 26-day EMAs.
- **Purpose:** It helps detect bullish and bearish trend changes via crossovers of the MACD line and signal line.
- **Application:** MACD was used to analyze potential buy/sell signals and confirm trend direction.

---

These indicators provided insights into the price behavior of Amazon and Google stocks, highlighting periods of strong trends, volatility spikes, and potential reversal points. This analysis serves as a foundation for the subsequent time series modeling and deeper performance comparison between the two stocks.

BOLLINGER BANDS

```python
#  Compute Bollinger Bands
def calculate_bollinger_bands(data, window=20):
    sma = data.rolling(window=window).mean()
    std = data.rolling(window=window).std()
    upper_band = sma + (std * 2)
    lower_band = sma - (std * 2)
    return sma, upper_band, lower_band

#  Amazon Bollinger Bands
amazon_cleaned['SMA_20'], amazon_cleaned['Upper_Band'], amazon_cleaned['Lower_Band'] = calculate_bollinger_bands(amazon_cleaned['Close_price'])

#  Google Bollinger Bands
google_cleaned['SMA_20'], google_cleaned['Upper_Band'], google_cleaned['Lower_Band'] = calculate_bollinger_bands(google_cleaned['Close_price'])

#  Plot Bollinger Bands
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # Amazon
plt.plot(amazon_cleaned['Date'], amazon_cleaned['Close_price'], label='Close Price', color='blue')
plt.plot(amazon_cleaned['Date'], amazon_cleaned['Upper_Band'], label='Upper Band', linestyle='dotted', color='red')
plt.plot(amazon_cleaned['Date'], amazon_cleaned['Lower_Band'], label='Lower Band', linestyle='dotted', color='green')
plt.legend()
plt.title("Amazon Bollinger Bands")

plt.subplot(1, 2, 2)  # Google
plt.plot(google_cleaned['Date'], google_cleaned['Close_price'], label='Close Price', color='green')
plt.plot(google_cleaned['Date'], google_cleaned['Upper_Band'], label='Upper Band', linestyle='dotted', color='red')
plt.plot(google_cleaned['Date'], google_cleaned['Lower_Band'], label='Lower Band', linestyle='dotted', color='blue')
plt.legend()
plt.title("Google Bollinger Bands")

plt.tight_layout()
plt.show()
```

```python
#  Compute RSI
def calculate_rsi(data, window=14):
    delta = data.diff(1)
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)

    avg_gain = pd.Series(gain).rolling(window=window, min_periods=1).mean()
    avg_loss = pd.Series(loss).rolling(window=window, min_periods=1).mean()

    rs = avg_gain / (avg_loss + 1e-10)  # Avoid division by zero
    rsi = 100 - (100 / (1 + rs))
    return rsi

amazon_cleaned['RSI_14'] = calculate_rsi(amazon_cleaned['Close_price'])
google_cleaned['RSI_14'] = calculate_rsi(google_cleaned['Close_price'])

#  Plot RSI
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # Amazon
plt.plot(amazon_cleaned['Date'], amazon_cleaned['RSI_14'], label='RSI', color='purple')
plt.axhline(70, linestyle='dashed', color='red')  # Overbought
plt.axhline(30, linestyle='dashed', color='green')  # Oversold
plt.legend()
plt.title("Amazon RSI")

plt.subplot(1, 2, 2)  # Google
plt.plot(google_cleaned['Date'], google_cleaned['RSI_14'], label='RSI', color='orange')
plt.axhline(70, linestyle='dashed', color='red')  # Overbought
plt.axhline(30, linestyle='dashed', color='green')  # Oversold
plt.legend()
plt.title("Google RSI")

plt.tight_layout()
plt.show()
```

MOVING AVERAGE CONVERGENCE DIVERGENCE

```python
#  Compute MACD
def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    short_ema = data.ewm(span=short_window, adjust=False).mean()
    long_ema = data.ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    return macd, signal

amazon_cleaned['MACD'], amazon_cleaned['Signal_Line'] = calculate_macd(amazon_cleaned['Close_price'])
google_cleaned['MACD'], google_cleaned['Signal_Line'] = calculate_macd(google_cleaned['Close_price'])

#  Plot MACD
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  # Amazon
plt.plot(amazon_cleaned['Date'], amazon_cleaned['MACD'], label='MACD', color='blue')
plt.plot(amazon_cleaned['Date'], amazon_cleaned['Signal_Line'], label='Signal Line', linestyle='dashed', color='red')
plt.legend()
plt.title("Amazon MACD")

plt.subplot(1, 2, 2)  # Google
plt.plot(google_cleaned['Date'], google_cleaned['MACD'], label='MACD', color='green')
plt.plot(google_cleaned['Date'], google_cleaned['Signal_Line'], label='Signal Line', linestyle='dashed', color='red')
plt.legend()
plt.title("Google MACD")

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro

# Calculate daily returns for Google and Amazon separately
google_returns = google_cleaned['Close_price'].pct_change().dropna()
amazon_returns = amazon_cleaned['Close_price'].pct_change().dropna()

# Perform Shapiro-Wilk normality test
shapiro_google = shapiro(google_returns)
shapiro_amazon = shapiro(amazon_returns)

# Print results
print(f"Google: W-statistic = {shapiro_google.statistic}, P-value = {shapiro_google.pvalue}")
print(f"Amazon: W-statistic = {shapiro_amazon.statistic}, P-value = {shapiro_amazon.pvalue}")
```

The Amazon W-statistic show that it is highly un-normal distribution compared to google 
Since the P-values are nearly zero then we confirm that neither google nor amazon follow a normal distribution

## TIMESERIES ANALYSIS

Time series analysis is a statistical technique that deals with time-ordered data points. In this report, time series analysis was applied to monthly closing price data for Amazon and Google stocks, with the goal of understanding trends, seasonality, stationarity, and making future forecasts.

The dataset used includes 13 months of cleaned historical stock prices, with a focus on the `Close_price` column. The key steps and methodologies implemented are detailed below.

---

### Step 1: Data Preparation and Resampling

**Objective:**  
To convert daily data into monthly average closing prices to reduce noise and identify meaningful trends.

**What was done:**  
- The `Close_price` column was resampled using monthly frequency.
- A new time series was generated for both Amazon and Google stock prices using the mean of each month.

---

### Step 2: Visualization of Monthly Trends

**Objective:**  
To visually inspect the historical behavior of stock prices over time.

**What was done:**  
- Line plots were generated for both Amazon and Google to display the monthly average closing prices.
- These plots helped identify upward or downward trends, volatility, and any visible cyclic patterns.

---

### Step 3: Stationarity Check Using Augmented Dickey-Fuller (ADF) Test

**Objective:**  
To determine whether the time series is stationary, which is a prerequisite for ARIMA modeling.

**What was done:**  
- The ADF test was applied to both Amazon and Google time series.
- The p-values for both series were above 0.05, indicating non-stationarity.
- First-order differencing was applied to both series to achieve stationarity.

---

### Step 4: Autocorrelation and Partial Autocorrelation Analysis

**Objective:**  
To identify appropriate parameters (lags) for ARIMA modeling using ACF and PACF plots.

**What was done:**  
- ACF and PACF plots were generated on the differenced data.
- The patterns observed in these plots guided the selection of ARIMA model parameters (p, d, q).

---

### Step 5: ARIMA Model Fitting

**Objective:**  
To model the time series data and prepare for forecasting.

**What was done:**  
- ARIMA models were fitted to the differenced, stationary series for both Amazon and Google.
- The models were trained on historical monthly closing price data.

---

### Step 6: Forecasting Future Stock Prices

**Objective:**  
To predict future stock prices for the next 3 months and visualize the forecasts.

**What was done:**  
- Forecasts were generated using the trained ARIMA models.
- Confidence intervals for the predictions were calculated.
- Forecast plots were created, showing historical trends, forecasted prices, and confidence intervals for both Amazon and Google.

---



Step 1: Data Preparation and Resampling

```python
# Step 1: Convert 'Date' column to datetime
amazon_cleaned['Date'] = pd.to_datetime(amazon_cleaned['Date'])
google_cleaned['Date'] = pd.to_datetime(google_cleaned['Date'])

# Step 2: Set 'Date' as the index
amazon_cleaned.set_index('Date', inplace=True)
google_cleaned.set_index('Date', inplace=True)

# Step 3: Resample to monthly average Close_price
amzn_monthly = amazon_cleaned['Close_price'].resample('M').mean()
goog_monthly = google_cleaned['Close_price'].resample('M').mean()
```

Step 2: Visualization of Monthly Trends

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(amzn_monthly, label='Amazon', color='orange')
plt.plot(goog_monthly, label='Google', color='blue')
plt.title('Monthly Average Closing Price: Amazon vs Google')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Monthly Average Closing Price: Amazon vs Google

This plot shows the **monthly average closing prices** of Amazon and Google over time, giving a broader view of stock performance and trends.

###  Observations:
- **Amazon** displays a **steadier upward trend** over the year, showing consistent growth with only minor setbacks.
- **Google** also trends upwards but with **greater fluctuations**, particularly mid-year.
- Despite these fluctuations, both companies show a **positive price trajectory** by the end of the observed period.
- The comparison suggests that **Amazon had more stable growth**, while **Google was more volatile**, though both ended strong.

###  Insight:
This visualization is useful for **performance comparison** between the two companies and complements the differenced graph by showing the actual price levels and trends before removing trends for modeling.


Step 3: Stationarity Check Using Augmented Dickey-Fuller (ADF) Test

```python
from statsmodels.tsa.stattools import adfuller

# Run ADF test for Amazon
result_amzn = adfuller(amzn_monthly)

print("Amazon ADF Statistic:", result_amzn[0])
print("Amazon p-value:", result_amzn[1])
```

```python
# Run ADF test for Google
result_goog = adfuller(goog_monthly)

print("Google ADF Statistic:", result_goog[0])
print("Google p-value:", result_goog[1])
```

```python
# First-order differencing
amzn_diff = amzn_monthly.diff().dropna()
goog_diff = goog_monthly.diff().dropna()

# Plot differenced series
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(amzn_diff)
plt.title("Amazon - First Differenced Monthly Close Price")

plt.subplot(1, 2, 2)
plt.plot(goog_diff)
plt.title("Google - First Differenced Monthly Close Price")

plt.tight_layout()
plt.show()
```

##  First Differenced Monthly Closing Price

The *first differenced* monthly closing price graphs for Amazon and Google illustrate the **month-to-month changes in stock prices**. This transformation is used to make the data **stationary**, which is a key requirement for many time series forecasting models like ARIMA.

###  Observations:
- The graphs show **no clear repeating patterns or trends**, which confirms **stationarity**.
- Both Amazon and Google exhibit **sudden jumps and drops** in the differenced values.
- The **volatility** in the graphs indicates that **past price changes do not predict future changes**, meaning the stock behavior is largely unpredictable in the short term.
- This is a **positive outcome** for forecasting models, as it ensures that the underlying data meets the assumptions required for accurate predictions.


Step 4: Autocorrelation and Partial Autocorrelation Analysis

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Amazon ACF & PACF
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(amzn_diff, ax=axes[0], lags=5)
plot_pacf(amzn_diff, ax=axes[1], lags=5)
axes[0].set_title('Amazon ACF')
axes[1].set_title('Amazon PACF')
plt.tight_layout()
plt.show()

# Google ACF & PACF
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(goog_diff, ax=axes[0], lags=5)
plot_pacf(goog_diff, ax=axes[1], lags=5)
axes[0].set_title('Google ACF')
axes[1].set_title('Google PACF')
plt.tight_layout()
plt.show()
```



The ACF and PACF plots show that both Amazon and Google stock prices do not have strong autocorrelations. This means:

- Price changes are mostly random and do not strongly depend on previous days.
- This is a sign of an efficient market where predicting future prices based on the past is difficult.

### Why this matters for the project:

- It creates a **realistic testing environment** for comparing model performance.
- Models cannot rely on clear trends — they are challenged to perform well despite randomness.
- This highlights the strength of each model in handling unpredictable behavior, making the comparison more meaningful.


Step 5: ARIMA Model Fitting

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA for Amazon
amzn_model = ARIMA(amzn_monthly, order=(1, 1, 1))
amzn_result = amzn_model.fit()
print("Amazon ARIMA Summary:")
print(amzn_result.summary())

# Fit ARIMA for Google
goog_model = ARIMA(goog_monthly, order=(1, 1, 1))
goog_result = goog_model.fit()
print("\nGoogle ARIMA Summary:")
print(goog_result.summary())
```

##  Amazon Stock Forecast using ARIMA

###  Model Used
I applied an **ARIMA(1,1,1)** model to forecast Amazon's stock closing prices:
- **AR(1)**: The model uses the most recent price to help predict the next.
- **I(1)**: The data is differenced once to remove trends and stabilize it.
- **MA(1)**: The model tries to correct for random fluctuations in price changes.

### Model Performance
- The model fits the data **reasonably well**.
- The **autoregressive (AR) term is significant**, indicating that past values have a strong effect on future prices.
- The **moving average (MA) term is not significant**, showing that recent noise is hard to model.
- Residuals (forecast errors) appear to be **normally distributed and uncorrelated**, which supports the model’s validity.

###  Forecast Summary
- The model forecasts that **Amazon's stock price may continue to rise** over the next 3 months.
- The **confidence interval is wide**, meaning there is uncertainty in the exact price range.
- This forecast reflects both the trend and potential variability due to market changes.

### Conclusion
This ARIMA(1,1,1) model provides a **basic but useful prediction** of Amazon’s stock behavior. It suggests upward momentum may continue, though the exact path depends on market volatility and external factors.

##  Google Stock Forecast using ARIMA

###  Model Used  
An **ARIMA(1,1,1)** model was applied to forecast Google’s stock closing prices:
- **AR(1)**: The model uses the most recent price to influence the next predicted value.
- **I(1)**: The data was differenced once to remove trend and make it stationary.
- **MA(1)**: The model adjusts for short-term shocks or random noise in the data.

### Model Performance  
- The model shows a **moderate fit** to the available data.
- **Neither the AR nor MA terms are statistically significant**, suggesting limited predictive strength from recent trends or noise.
- The **residual variance (sigma²)** is significant, meaning there's measurable fluctuation in the data.
- The **Ljung-Box test** indicates that residuals are **not autocorrelated**, which is a good sign.
- However, the **Jarque-Bera test** suggests that residuals **are not normally distributed**, and **skewness and kurtosis** indicate heavy tails and asymmetry.
- The model was trained on **only 13 observations**, which **limits its reliability**.

###  Forecast Summary  
- The forecast shows a **relatively flat trend**, suggesting that Google's stock price may **stabilize near 200 USD** over the next 3 months.
- The **confidence interval is wide** (roughly from 160 to 240 USD), indicating **uncertainty in direction**.
- The lack of significant coefficients and non-normal residuals imply that this model should be **interpreted with caution**.

Step 6: Forecasting Future Stock Prices

```python
import matplotlib.pyplot as plt

# Forecast the next 3 months
forecast_amzn = amzn_result.get_forecast(steps=3)
forecast_goog = goog_result.get_forecast(steps=3)

# Get confidence intervals
conf_int_amzn = forecast_amzn.conf_int()
conf_int_goog = forecast_goog.conf_int()

# Forecasted mean values
forecast_amzn_mean = forecast_amzn.predicted_mean
forecast_goog_mean = forecast_goog.predicted_mean

# Plot Amazon forecast
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(amzn_monthly, label='Historical (Amazon)')
plt.plot(forecast_amzn_mean, label='Forecast (Amazon)', color='green')
plt.fill_between(conf_int_amzn.index,
                 conf_int_amzn.iloc[:, 0],
                 conf_int_amzn.iloc[:, 1],
                 color='green', alpha=0.2)
plt.title('Amazon 3-Month Forecast')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()

# Plot Google forecast
plt.subplot(1, 2, 2)
plt.plot(goog_monthly, label='Historical (Google)')
plt.plot(forecast_goog_mean, label='Forecast (Google)', color='orange')
plt.fill_between(conf_int_goog.index,
                 conf_int_goog.iloc[:, 0],
                 conf_int_goog.iloc[:, 1],
                 color='orange', alpha=0.2)
plt.title('Google 3-Month Forecast')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()

plt.tight_layout()
plt.show()
```
